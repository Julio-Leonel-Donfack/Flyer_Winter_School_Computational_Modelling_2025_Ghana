### Exercise 1: Solving a Simple Differential Equation

**Problem Statement**: Solve the following differential equation using Euler's method:

$$
\frac{dy}{dt} = -ky
$$

Where \( k = 0.1 \) and \( y(0) = 10 \).

#### Tasks:
1. Implement Euler's method to approximate the solution.
2. Visualize the solution for different values of \( k \).
3. Interpret the physical meaning of this equation in the context of the degradation of a substance in the ocean (e.g., pollution dissipating in water).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 0.1
y0 = 10
t_max = 50
dt = 0.1
time = np.arange(0, t_max, dt)

# Euler's Method
y = np.zeros_like(time)
y[0] = y0

for i in range(1, len(time)):
    y[i] = y[i-1] - k * y[i-1] * dt

# Visualization
plt.plot(time, y)
plt.title("Solution of the Differential Equation (Euler's Method)")
plt.xlabel("Time (t)")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```

```python
## 1. Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt

## 2. Defining Euler's method

def euler_method(k, y0, t_max, h):
    # Number of iterations
    N = int(t_max / h)
    
    # Initializing arrays
    t_values = np.zeros(N)
    y_values = np.zeros(N)
    
    # Initial conditions
    t_values[0] = 0
    y_values[0] = y0
    
    # Applying Euler's method
    for n in range(1, N):
        t_values[n] = t_values[n-1] + h
        y_values[n] = y_values[n-1] - h * k * y_values[n-1]  # -k*y for the differential equation
        
    return t_values, y_values

## 3. Visualizing the solution for different values of k

# Parameters
y0 = 10    # Initial condition
t_max = 50  # Maximum time
h = 0.1    # Time step

# Values of k to test
k_values = [0.05, 0.1, 0.2]

# Plot
plt.figure(figsize=(10, 6))

for k in k_values:
    t, y = euler_method(k, y0, t_max, h)
    plt.plot(t, y, label=f'k = {k}')

plt.title('Solution of the Differential Equation with Euler\'s Method')
plt.xlabel('Time (t)')
plt.ylabel('Concentration (y)')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. Physical Interpretation of the Equation

The differential equation

$$
\frac{dy}{dt} = -ky
$$

models a degradation or dissipation process of a substance in a medium. In this particular case, it can be related to the **degradation of pollution in the ocean**.

#### Physical Significance:
- \( y(t) \) represents the concentration of the substance (e.g., pollution) at a given time \( t \).
- \( k \) is a constant degradation rate. It indicates how quickly the substance degrades or dissipates. The larger the value of \( k \), the faster the substance degrades.
- The negative sign in the equation \(\frac{dy}{dt} = -ky\) shows that the concentration of the substance decreases over time.

When \( k \) is small, the degradation of the substance is slow, and the concentration remains high for a longer period. Conversely, when \( k \) is large, the degradation is rapid, and the concentration drops quickly.

### 5. Conclusion

Euler's method allows us to approximately solve this differential equation. Depending on the value of \( k \), the behavior of the substance's concentration in the ocean changes, illustrating the importance of the **degradation rate** in modeling the dissipation of pollution in a natural environment.


# Exercise 2: Modeling the Growth of a Fish Population

## Problem Statement:
Use the **logistic growth differential equation** to model the fish population in a marine ecosystem:

$$
\frac{dP}{dt} = rP \left( 1 - \frac{P}{K} \right)
$$

Where:
- \( P(t) \) is the fish population at time \( t \),
- \( r \) is the growth rate,
- \( K \) is the maximum carrying capacity of the environment (the maximum population the ecosystem can sustain).

## Tasks:
1. Implement this equation and solve numerically for the following values:
    - \( P(0) = 10 \),
    - \( r = 0.2 \),
    - \( K = 1000 \).

2. Visualize the evolution of the population over time.

3. Discuss the impact of increasing or decreasing the growth rate \( r \) on the population.

```python
# Parameters of the marine ecosystem
r = 0.2  # Growth rate
K = 1000  # Carrying capacity
P0 = 10  # Initial population
t_max = 100
dt = 0.1
time = np.arange(0, t_max, dt)

# Euler's method for logistic growth
P = np.zeros_like(time)
P[0] = P0

for i in range(1, len(time)):
    P[i] = P[i-1] + r * P[i-1] * (1 - P[i-1] / K) * dt

# Visualization
plt.plot(time, P)
plt.title("Logistic Growth of the Fish Population")
plt.xlabel("Time (t)")
plt.ylabel("Population (P)")
plt.grid(True)
plt.show()
```

```python
# Cell 1: Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Cell 2: Defining parameters and the logistic growth function

# Parameters
r = 0.2     # Growth rate
K = 1000    # Carrying capacity of the environment
P0 = 10     # Initial population
t_max = 100 # Maximum simulation time
dt = 0.1    # Time step for the simulation

# Function representing the logistic growth differential equation
def growth_logistique(P, r, K):
    return r * P * (1 - P / K)

# Cell 3: Euler's method to solve the ODE
# Initializing lists for population values and time
times = np.arange(0, t_max, dt)
populations = np.zeros(len(times))
populations[0] = P0

# Euler's method to solve the ODE
for i in range(1, len(times)):
    populations[i] = populations[i-1] + growth_logistique(populations[i-1], r, K) * dt

# Cell 4: Visualizing the population evolution
plt.plot(times, populations)
plt.title("Fish Population Evolution")
plt.xlabel("Time (t)")
plt.ylabel("Population (P)")
plt.grid(True)
plt.show()

# Cell 5: Testing the impact of different growth rates (r)

# New function to test with different growth rates
def simulate_population(r):
    populations = np.zeros(len(times))
    populations[0] = P0
    for i in range(1, len(times)):
        populations[i] = populations[i-1] + growth_logistique(populations[i-1], r, K) * dt
    return populations

# Different growth rates to test
r_values = [0.2, 0.5, 0.05]

# Visualizing the results for multiple r values
plt.figure(figsize=(10, 6))
for r in r_values:
    populations = simulate_population(r)
    plt.plot(times, populations, label=f'r = {r}')
    
plt.title("Impact of Growth Rate on Population Evolution")
plt.xlabel("Time (t)")
plt.ylabel("Population (P)")
plt.legend()
plt.grid(True)
plt.show()
```

# Conclusion and Interpretation of Results

1. **For r = 0.2:** The population grows moderately and gradually reaches the maximum capacity \( K \).
2. **For r = 0.5:** The population quickly reaches \( K \), showing a faster exponential growth before stabilizing.
3. **For r = 0.05:** The population grows slowly, taking much longer to reach the carrying capacity \( K \), and may not reach it within a reasonable time.

### General Conclusion:
The logistic growth model helps us understand how the carrying capacity of an environment limits population growth. The growth rate \( r \) plays a crucial role in this dynamic. A high \( r \) causes rapid saturation of the population at the capacity \( K \), while a low \( r \) results in slower growth.

# Tasks for Modeling a School of Fish with ABM

To model a school of fish using agent-based modeling (ABM), we will follow several key steps. This project includes creating agents (fish), adding a predator, defining behavior rules, and simulating and visualizing interactions. Below is a detailed guide for each task.

## 1. Creating the Agents (Fish)

### a. Defining Agent Properties

Each fish can have the following properties:
- **Position**: Coordinates (x, y) in the space.
- **Speed**: The movement speed.
- **Direction**: The angle of movement.
- **Size**: The size of the fish (which may influence interactions).

### b. Behavior Rules

The fish must follow simple rules:
- **Cohesion**: Move towards the center of the school.
  - Calculate the center of mass of the group and adjust direction towards that point.
- **Separation**: Avoid collisions.
  - If another fish is too close, change direction to avoid a collision.
- **Alignment**: Align their direction with their neighbors.
  - Observe the direction of neighboring fish and adjust their own direction accordingly.

## 2. Adding a Predator

### a. Defining the Predator

The predator can also be an agent with its own properties:
- **Position**: Coordinates (x, y).
- **Speed**: The predator’s speed is higher than that of the fish.
- **Direction**: The angle of movement.

### b. Hunting Rules

The rules for the predator include:
- **Fish Tracking**: Move towards the closest fish.
- **Hunting**: If the predator is within a certain distance of a fish, it "attacks" (simulates a capture).
- **Avoiding Fish**: Fish must react to the predator by quickly moving away when it is nearby.

## 3. Simulation and Visualization

### a. Simulation Environment

Use a simulation environment such as:
- **Processing**: Ideal for simple graphical visualizations.
- **NetLogo**: Specifically designed for ABM, with an intuitive user interface.
- **Python with Matplotlib or Pygame**: For a more customized approach.

### b. Implementing the Simulation

- **Initializing the Agents**:
  - Create a certain number of fish and a predator with random positions in space.
  
- **Simulation Loop**:
  - At each iteration, update the position and direction of each fish according to the defined rules.
  - Update the position of the predator according to its hunting rules.
  - Check for collisions between the predator and the fish.

- **Visualization**:
  - Display the fish and the predator on the screen, updating their position at each iteration.
  - Use different colors or shapes to distinguish the fish from the predator.

```python
import pygame
import numpy as np
import random

# Simulation parameters
WIDTH, HEIGHT = 800, 600
NUM_FISH = 50
NUM_PREDATOR = 3
SPEED = 2
AVOID_RADIUS = 30
ALIGN_RADIUS = 50
CENTER_RADIUS = 200

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Fish School Simulation")

# Classes
class Fish:
    def __init__(self, x, y):
        self.pos = np.array([x, y], dtype=np.float64)  # Ensure position is in float64
        self.vel = np.random.rand(2) * 2 - 1  # Random velocity
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED

    def move(self, fishes, predator):
        center_of_mass = np.zeros(2)
        avg_velocity = np.zeros(2)
        avoid_force = np.zeros(2)

        # Interaction rules with other fish
        for fish in fishes:
            if np.linalg.norm(fish.pos - self.pos) < AVOID_RADIUS:
                avoid_force += self.pos - fish.pos
            if np.linalg.norm(fish.pos - self.pos) < ALIGN_RADIUS:
                avg_velocity += fish.vel
            if np.linalg.norm(fish.pos - self.pos) < CENTER_RADIUS:
                center_of_mass += fish.pos

        # Rule for the center of the school
        center_of_mass /= len(fishes)
        move_towards_center = center_of_mass - self.pos

        # Combine forces
        self.vel += (move_towards_center * 0.05 + avg_velocity * 0.1 + avoid_force * 0.1)
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED  # Normalize the velocity

        # Move
        self.pos += self.vel

        # Wrap around the window
        self.pos = self.pos % np.array([WIDTH, HEIGHT])

    def draw(self):
        # Convert position to int for drawing
        pygame.draw.circle(screen, (0, 255, 0), self.pos.astype(int), 5)

class Predator:
    def __init__(self, x, y):
        self.pos = np.array([x, y], dtype=np.float64)
        self.vel = np.random.rand(2) * 2 - 1
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED * 1.5  # Predator is faster

    def move(self, fishes):
        closest_fish = None
        min_dist = float('inf')

        # Find the closest fish
        for fish in fishes:
            dist = np.linalg.norm(fish.pos - self.pos)
            if dist < min_dist:
                min_dist = dist
                closest_fish = fish

        # Move towards the closest fish
        if closest_fish:
            move_towards_fish = closest_fish.pos - self.pos
            self.vel = move_towards_fish / np.linalg.norm(move_towards_fish) * SPEED * 1.5

        # Move
        self.pos += self.vel
        self.pos = self.pos % np.array([WIDTH, HEIGHT])

    def draw(self):
        # Convert position to int for drawing
        pygame.draw.circle(screen, (255, 0, 0), self.pos.astype(int), 10)

# Initialize fish and predator
fishes = [Fish(random.randint(0, WIDTH), random.randint(0, HEIGHT)) for _ in range(NUM_FISH)]
predator = Predator(random.randint(0, WIDTH), random.randint(0, HEIGHT))

# Simulation loop
running = True
while running:
    screen.fill((0, 0, 0))  # Black background
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the fish
    for fish in fishes:
        fish.move(fishes, predator)

    # Move the predator
    predator.move(fishes)

    # Draw the fish
    for fish in fishes:
        fish.draw()

    # Draw the predator
    predator.draw()

    # Update the display
    pygame.display.flip()
    pygame.time.Clock().tick(30)

pygame.quit()
```

# Coral Reef Ecosystem Modeling

## 1. Creation of Agents:

**Properties:**
- **Population:** Number of herbivores present.

## 2. Definition of Interactions:

### a. Interaction Between Herbivores and Algae:
Herbivores consume algae. If herbivores eat too much algae, it reduces the algae population, which may harm coral regeneration. The rate at which herbivores consume algae may depend on factors such as herbivore population density or the available amount of algae.

### b. Interaction Between Algae and Corals:
If there are enough algae, corals regenerate more rapidly. If algae are scarce (due to overconsumption by herbivores or other factors), corals suffer and may die. The regeneration rate of corals depends on the abundance of algae and the intensity of the pressure from herbivores.

### c. Herbivore Pressure on Corals:
If there are too many herbivores or if they eat too much algae, it can lead to overconsumption of resources, preventing corals from regenerating sufficiently, which could lead to their decline.

## 3. Modeling Dynamics:

### a. Algae Growth Model:
Algae grow at a rate \( G_{\text{algae}} \) based on available resources and water temperature. They are consumed at a rate \( C_{\text{herbivores}} \) by herbivores. A simple equation for the amount of algae \( A(t) \) at time \( t \) could be:
$$
\frac{dA}{dt} = G_{\text{algae}} - C_{\text{herbivores}} \times A(t) \times P_{\text{herbivores}}(t)
$$

### b. Coral Regeneration Model:
Corals regenerate at a rate \( R_{\text{corals}} \) based on available algae, but this decreases if herbivores consume too many algae. The equation for coral health \( S_{\text{corals}}(t) \) could be:
$$
\frac{dS_{\text{corals}}}{dt} = R_{\text{corals}} \times A(t) - \text{herbivore pressure}
$$

### c. Herbivore Population Dynamics:
The herbivore population \( P_{\text{herbivores}}(t) \) varies depending on the amount of algae available and water temperature. A possible dynamic is:
$$
\frac{dP_{\text{herbivores}}}{dt} = r \times P_{\text{herbivores}}(t) \times A(t) - d \times P_{\text{herbivores}}(t)
$$
where:
- \( r \) is the reproduction rate of herbivores.
- \( d \) is the mortality rate of herbivores.
- \( A(t) \) is the amount of algae available, influencing herbivore population growth.

## 4. Scenarios to Simulate:

- **Scenario 1:** Increase in herbivore population. Observe the impact on algae and coral health as herbivores consume more algae.
- **Scenario 2:** Increase in water temperature. Rising temperatures could negatively affect the growth of corals and algae, altering the ecological balance.
- **Scenario 3:** Decrease in herbivore population. Algae may regenerate more easily, allowing corals to stabilize.

## 5. Observation and Analysis:
For each scenario, the evolution of algae populations, coral health, and herbivore populations can be observed over time. The simulation can be performed over a discrete time interval (e.g., monthly or yearly) to observe long-term changes. The results can be visualized using graphs that show the evolution of populations and coral health under the influence of different parameters.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
G_algae, C_herbivores, R_corals, Temp_rate, Coral_health_threshold = 0.1, 0.05, 0.1, 1.0, 0.2
r_herbivores, d_herbivores = 0.1, 0.05

def model(y, t, G_algae, C_herbivores, R_corals, r_herbivores, d_herbivores):
    A, S_corals, P_herbivores = y
    dA_dt = G_algae - C_herbivores * P_herbivores * A
    herbivore_pressure = 0.05 * P_herbivores
    dS_corals_dt = R_corals * A - herbivore_pressure
    dP_herbivores_dt = r_herbivores * P_herbivores * A - d_herbivores * P_herbivores
    return [dA_dt, dS_corals_dt, dP_herbivores_dt]

A0, S0, P0 = 0.5, 1.0, 50
y0 = [A0, S0, P0]
t = np.linspace(0, 100, 1000)
params = (G_algae, C_herbivores, R_corals, r_herbivores, d_herbivores)
solution = odeint(model, y0, t, args=params)

A, S_corals, P_herbivores = solution[:, 0], solution[:, 1], solution[:, 2]

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.plot(t, A, color='green')
plt.xlabel('Time')
plt.ylabel('Algae Abundance')
plt.title('Algae Dynamics')
plt.grid(True)
plt.subplot(1, 3, 2)
plt.plot(t, S_corals, color='blue')
plt.xlabel('Time')
plt.ylabel('Coral Health')
plt.title('Coral Health Evolution')
plt.grid(True)
plt.subplot(1, 3, 3)
plt.plot(t, P_herbivores, color='orange')
plt.xlabel('Time')
plt.ylabel('Herbivore Population')
plt.title('Herbivore Dynamics')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Simulation parameters ---
timestep = 1  # One time unit (e.g., years)
num_years = 100  # Duration of the simulation (in years)

# Climate change parameters
initial_pH = 8.1  # Initial water pH
acidification_rate = 0.01  # Change in pH per year (acidification)

# Initial populations
initial_phytoplankton = 1000  # Initial phytoplankton population
initial_corals = 500  # Initial coral population
initial_fish = 300  # Initial fish population

# --- Classes for agents ---
# 1. Phytoplankton: Grows based on pH and nutrient availability.
class Phytoplankton:
    def __init__(self, population):
        self.population = population

    def grow(self, pH):
        # Phytoplankton grows more slowly at lower pH
        growth_rate = 0.1 if pH > 7.8 else 0.05  # Slower growth if acidification progresses
        self.population += growth_rate * self.population

# 2. Corals: Depend on pH to grow and reproduce.
class Coral:
    def __init__(self, population):
        self.population = population

    def grow(self, pH):
        # Coral growth depending on pH
        if pH < 7.8:
            death_rate = 0.1  # Increased death rate with acidification
        else:
            death_rate = 0.02  # Normal growth
        self.population *= (1 - death_rate)

# 3. Fish: Feed on phytoplankton and are sensitive to pH decrease.
class Fish:
    def __init__(self, population):
        self.population = population

    def feed_and_reproduce(self, phytoplankton_population, pH):
        # Fish feed on phytoplankton and reproduce based on its availability
        if pH < 7.5:  # Stress due to acidification
            survival_rate = 0.5  # Less reproduction and higher mortality
        else:
            survival_rate = 0.9  # Higher survival rate in better conditions
        # Fish eat phytoplankton, so their population depends on this food source
        self.population += (0.05 * phytoplankton_population * survival_rate)

# --- Initializing the agents ---
# Initializing the population of each species
phytoplankton = Phytoplankton(initial_phytoplankton)
corals = Coral(initial_corals)
fish = Fish(initial_fish)

# List to track results over the years
results = []

# --- Simulation over several years ---
for year in range(num_years):
    # Calculate pH each year (with acidification)
    pH = initial_pH - acidification_rate * year
    
    # Agents interact according to the current pH
    phytoplankton.grow(pH)
    corals.grow(pH)
    fish.feed_and_reproduce(phytoplankton.population, pH)
    
    # Record the current state of the ecosystem
    results.append([year, pH, phytoplankton.population, corals.population, fish.population])

# --- Convert results to DataFrame ---
df = pd.DataFrame(results, columns=["Year", "pH", "Phytoplankton", "Corals", "Fish"])
df.set_index("Year", inplace=True)

# Display the first rows of the table to verify the results
df.head()

# --- Visualization of results ---

# Create a figure to display the graphs
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Graph 1: Evolution of water pH
ax[0].plot(df.index, df["pH"], color="blue", label="pH")
ax[0].set_title("Evolution of Ocean pH")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("pH")
ax[0].grid(True)

# Graph 2: Evolution of Phytoplankton, Corals, and Fish populations
ax[1].plot(df.index, df["Phytoplankton"], label="Phytoplankton", color="green")
ax[1].plot(df.index, df["Corals"], label="Corals", color="orange")
ax[1].plot(df.index, df["Fish"], label="Fish", color="red")
ax[1].set_title("Evolution of Marine Populations")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Population")
ax[1].legend()
ax[1].grid(True)

# Adjust layout for better presentation
plt.tight_layout()
plt.show()
```

- A plot showing the **progressive decrease in pH** of the ocean.
- A plot illustrating the **evolution of populations**:
  - Phytoplankton might grow initially, but its population could later decline as the pH drops.
  - Corals should decrease more significantly due to their dependence on pH for growth.
  - Fish will suffer the consequences of reduced food (phytoplankton) and the stress caused by acidification.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Simulation Parameters ---
timestep = 1  # Time unit (e.g., years)
num_years = 100  # Duration of the simulation (in years)

# SST (Sea Surface Temperature) parameters
initial_SST = 25  # Initial sea surface temperature (in °C)
warming_rate = 0.1  # Annual temperature increase (in °C per year)

# Initial populations
initial_phytoplankton = 1000  # Initial number of phytoplankton
initial_corals = 500  # Initial number of corals
initial_fish = 300  # Initial number of fish

# --- Agent Classes ---
# 1. Phytoplankton: Grows based on sea surface temperature.
class Phytoplankton:
    def __init__(self, population):
        self.population = population

    def grow(self, SST):
        # Phytoplankton grows best at moderate temperatures
        if SST > 28:
            growth_rate = -0.05  # Growth slows down if SST is too high
        elif SST < 18:
            growth_rate = 0.05  # Growth optimal at moderate temperatures
        else:
            growth_rate = 0.1  # Optimal growth at moderate temperatures (21-28°C)
        self.population += growth_rate * self.population

# 2. Coral: Corals experience increased stress with high temperatures (bleaching).
class Coral:
    def __init__(self, population):
        self.population = population

    def grow(self, SST):
        # Corals experience stress if SST exceeds a certain threshold
        if SST > 30:
            death_rate = 0.1  # Coral bleaching (loss of population)
        elif SST > 28:
            death_rate = 0.05  # Mild stress
        else:
            death_rate = 0.02  # Normal growth
        self.population *= (1 - death_rate)

# 3. Fish: Depend on phytoplankton availability and are sensitive to temperature.
class Fish:
    def __init__(self, population):
        self.population = population

    def feed_and_reproduce(self, phytoplankton_population, SST):
        # Fish suffer with high temperatures
        if SST > 30:  # Intense stress with high SST
            survival_rate = 0.5  # Less reproduction and higher mortality
        else:
            survival_rate = 0.9  # Normal growth
        # Fish feed on phytoplankton, so their population depends on this resource
        self.population += (0.05 * phytoplankton_population * survival_rate)

# --- Initialize the agents ---
# Initialize the populations of each species
phytoplankton = Phytoplankton(initial_phytoplankton)
corals = Coral(initial_corals)
fish = Fish(initial_fish)

# List to track the results over the years
results = []

# --- Simulation over multiple years ---
for year in range(num_years):
    # Calculate the SST for each year (with annual warming)
    SST = initial_SST + warming_rate * year
    
    # Agents interact based on the current SST
    phytoplankton.grow(SST)
    corals.grow(SST)
    fish.feed_and_reproduce(phytoplankton.population, SST)
    
    # Record the current state of the ecosystem
    results.append([year, SST, phytoplankton.population, corals.population, fish.population])

# --- Convert results to DataFrame ---
df = pd.DataFrame(results, columns=["Year", "SST", "Phytoplankton", "Corals", "Fish"])
df.set_index("Year", inplace=True)

# Display the first rows of the table to verify results
df.head()

# --- Visualize the results ---

# Create a figure to display the plots
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Graph 1: Evolution of SST
ax[0].plot(df.index, df["SST"], color="red", label="SST (°C)")
ax[0].set_title("Evolution of Sea Surface Temperature (SST)")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("SST (°C)")
ax[0].grid(True)

# Graph 2: Evolution of Phytoplankton, Coral, and Fish populations
ax[1].plot(df.index, df["Phytoplankton"], label="Phytoplankton", color="green")
ax[1].plot(df.index, df["Corals"], label="Corals", color="orange")
ax[1].plot(df.index, df["Fish"], label="Fish", color="blue")
ax[1].set_title("Evolution of Marine Populations")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Population")
ax[1].legend()
ax[1].grid(True)

# Adjust layout for better presentation
plt.tight_layout()
plt.show()
```
