### Exercice 1 : Résolution d'une Équation Différentielle Simple

**Énoncé** : Résoudre l'équation différentielle suivante en utilisant la méthode d'Euler :

$$
\frac{dy}{dt} = -ky
$$

Où \( k = 0.1 \) et \( y(0) = 10 \).

#### Tâches :
1. Implémenter la méthode d'Euler pour approximer la solution.
2. Visualiser la solution pour différentes valeurs de \( k \).
3. Interpréter la signification physique de cette équation dans le contexte de la dégradation d'une substance dans l'océan (ex : pollution qui se dissipe dans l'eau).

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
k = 0.1
y0 = 10
t_max = 50
dt = 0.1
time = np.arange(0, t_max, dt)

# Méthode d'Euler
y = np.zeros_like(time)
y[0] = y0

for i in range(1, len(time)):
    y[i] = y[i-1] - k * y[i-1] * dt

# Visualisation
plt.plot(time, y)
plt.title("Solution de l'équation différentielle (Méthode d'Euler)")
plt.xlabel("Temps (t)")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```

```python
## 1. Importation des bibliothèques nécessaires

import numpy as np
import matplotlib.pyplot as plt

##2. Définition de la méthode d'Euler

def euler_method(k, y0, t_max, h):
    # Nombre d'itérations
    N = int(t_max / h)
    
    # Initialisation des tableaux
    t_values = np.zeros(N)
    y_values = np.zeros(N)
    
    # Conditions initiales
    t_values[0] = 0
    y_values[0] = y0
    
    # Application de la méthode d'Euler
    for n in range(1, N):
        t_values[n] = t_values[n-1] + h
        y_values[n] = y_values[n-1] - h * k * y_values[n-1]  # -k*y pour l'équation différentielle
        
    return t_values, y_values

##3. Visualisation de la solution pour différentes valeurs de k

# Paramètres
y0 = 10    # Condition initiale
t_max = 50  # Temps maximal
h = 0.1    # Pas de temps

# Valeurs de k à tester
k_values = [0.05, 0.1, 0.2]

# Graphique
plt.figure(figsize=(10, 6))

for k in k_values:
    t, y = euler_method(k, y0, t_max, h)
    plt.plot(t, y, label=f'k = {k}')

plt.title('Solution de l\'équation différentielle avec la méthode d\'Euler')
plt.xlabel('Temps (t)')
plt.ylabel('Concentration (y)')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. Interprétation physique de l’équation

L'équation différentielle

$$
\frac{dy}{dt} = -ky
$$

modélise un phénomène de dégradation ou de dissipation d'une substance dans un milieu. Dans ce cas particulier, on peut la relier à la **dégradation de la pollution dans l'océan**.

#### Signification physique :
- \( y(t) \) représente la concentration de la substance (par exemple, la pollution) à un instant donné \( t \).
- \( k \) est un taux de dégradation constant. Il indique à quelle vitesse la substance se dégrade ou se dissipe. Plus \( k \) est grand, plus la substance se dégrade rapidement.
- Le signe négatif dans l'équation \(\frac{dy}{dt} = -ky\) montre que la concentration de la substance diminue au fil du temps.

Lorsque \( k \) est petit, la dégradation de la substance est lente et la concentration reste élevée pendant plus longtemps. À l'inverse, lorsque \( k \) est grand, la dégradation est rapide et la concentration chute rapidement.

### 5. Conclusion

La méthode d'Euler nous permet de résoudre approximativement cette équation différentielle. En fonction de la valeur de \( k \), le comportement de la concentration de la substance dans l'océan change, ce qui illustre l'importance de la **vitesse de dégradation** pour modéliser le phénomène de dissipation d'une pollution dans un environnement naturel.

# Exercice 2 : Modélisation de la croissance d’une population de poissons

## Énoncé :
Utilisez l’équation différentielle de **croissance logistique** pour modéliser la population de poissons dans un écosystème marin :

$$
\frac{dP}{dt} = rP \left( 1 - \frac{P}{K} \right)
$$

Où :
- \( P(t) \) est la population de poissons à l'instant \( t \),
- \( r \) est le taux de croissance,
- \( K \) est la capacité maximale du milieu (la population maximale que l'écosystème peut soutenir).

## Tâches :
1. Implémenter cette équation et résoudre numériquement pour les valeurs suivantes :
    - \( P(0) = 10 \),
    - \( r = 0.2 \),
    - \( K = 1000 \).

2. Visualiser l’évolution de la population en fonction du temps.

3. Discuter de l'impact de l'augmentation ou de la diminution du taux de croissance \( r \) sur la population.

```python
# Paramètres de l'écosystème marin
r = 0.2  # Taux de croissance
K = 1000  # Capacité maximale
P0 = 10  # Population initiale
t_max = 100
dt = 0.1
time = np.arange(0, t_max, dt)

# Méthode d'Euler pour la croissance logistique
P = np.zeros_like(time)
P[0] = P0

for i in range(1, len(time)):
    P[i] = P[i-1] + r * P[i-1] * (1 - P[i-1] / K) * dt

# Visualisation
plt.plot(time, P)
plt.title("Croissance logistique de la population de poissons")
plt.xlabel("Temps (t)")
plt.ylabel("Population (P)")
plt.grid(True)
plt.show()
```

```python
# Cellule 1 : Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt

# Cellule 2 : Définition des paramètres et de la fonction de croissance logistique

# Paramètres
r = 0.2     # Taux de croissance
K = 1000    # Capacité maximale de l'environnement
P0 = 10     # Population initiale
t_max = 100 # Temps maximal de simulation
dt = 0.1    # Pas de temps pour la simulation

# Fonction qui représente l'équation différentielle de croissance logistique
def growth_logistique(P, r, K):
    return r * P * (1 - P / K)

# Cellule 3 : Méthode d'Euler pour résoudre l'EDO
# Initialisation des listes pour les valeurs de P et du temps
times = np.arange(0, t_max, dt)
populations = np.zeros(len(times))
populations[0] = P0

# Méthode d'Euler pour résoudre l'EDO
for i in range(1, len(times)):
    populations[i] = populations[i-1] + growth_logistique(populations[i-1], r, K) * dt

# Cellule 4 : Visualisation de l'évolution de la population
plt.plot(times, populations)
plt.title("Évolution de la population de poissons")
plt.xlabel("Temps (t)")
plt.ylabel("Population (P)")
plt.grid(True)
plt.show()

# Cellule 5 : Tester l'impact de différents taux de croissance (r)

# Nouvelle fonction pour tester avec différents taux de croissance
def simulate_population(r):
    populations = np.zeros(len(times))
    populations[0] = P0
    for i in range(1, len(times)):
        populations[i] = populations[i-1] + growth_logistique(populations[i-1], r, K) * dt
    return populations

# Taux de croissance différents à tester
r_values = [0.2, 0.5, 0.05]

# Visualisation des résultats pour plusieurs valeurs de r
plt.figure(figsize=(10, 6))
for r in r_values:
    populations = simulate_population(r)
    plt.plot(times, populations, label=f'r = {r}')
    
plt.title("Impact du taux de croissance sur l'évolution de la population")
plt.xlabel("Temps (t)")
plt.ylabel("Population (P)")
plt.legend()
plt.grid(True)
plt.show()
```

# Tâches pour la Modélisation d'un Banc de Poissons avec ABM

Pour modéliser un banc de poissons en utilisant la modélisation basée sur des agents (ABM), nous allons suivre plusieurs étapes clés. Ce projet inclut la création d'agents (poissons), l'ajout d'un prédateur, la définition de règles de comportement, et la simulation et visualisation des interactions. Voici un guide détaillé pour chaque tâche.

## 1. Création des Agents (Poissons)

### a. Définition des Propriétés des Agents

Chaque poisson peut avoir les propriétés suivantes :
- **Position** : Coordonnées (x, y) dans l'espace.
- **Vitesse** : Vitesse de déplacement.
- **Direction** : Angle de déplacement.
- **Taille** : Taille du poisson (peut influencer les interactions).

### b. Règles de Comportement

Les poissons doivent suivre des règles simples :
- **Cohésion** : Se déplacer vers le centre du banc.
  - Calculer le centre de masse du groupe et ajuster la direction vers ce point.
- **Séparation** : Éviter les collisions.
  - Si un autre poisson est trop proche, changer de direction pour éviter une collision.
- **Alignement** : Aligner leur direction avec celle de leurs voisins.
  - Observer la direction des poissons voisins et ajuster leur propre direction en conséquence.

## 2. Ajout d'un Prédateur

### a. Définition du Prédateur

Le prédateur peut également être un agent avec ses propres propriétés :
- **Position** : Coordonnées (x, y).
- **Vitesse** : Vitesse de déplacement plus élevée que celle des poissons.
- **Direction** : Angle de déplacement.

### b. Règles de Chasse

Les règles pour le prédateur incluent :
- **Suivi des Poissons** : Se déplacer vers le poisson le plus proche.
- **Chasse** : Si le prédateur est à une certaine distance d'un poisson, il "attaque" (simule une capture).
- **Évitement des Poissons** : Les poissons doivent réagir au prédateur en s'éloignant rapidement lorsqu'il est à proximité.

## 3. Simulation et Visualisation

### a. Environnement de Simulation

Utiliser un environnement de simulation comme :
- **Processing** : Idéal pour les visualisations graphiques simples.
- **NetLogo** : Conçu spécifiquement pour l'ABM, avec une interface utilisateur intuitive.
- **Python avec Matplotlib ou Pygame** : Pour une approche plus personnalisée.

### b. Implémentation de la Simulation

- **Initialiser les Agents** :
  - Créer un certain nombre de poissons et un prédateur avec des positions aléatoires dans l'espace.
  
- **Boucle de Simulation** :
  - À chaque itération, mettre à jour la position et la direction de chaque poisson selon les règles définies.
  - Mettre à jour la position du prédateur selon ses règles de chasse.
  - Vérifier les collisions entre le prédateur et les poissons.

- **Visualisation** :
  - Afficher les poissons et le prédateur sur l'écran, en mettant à jour leur position à chaque itération.
  - Utiliser différentes couleurs ou formes pour distinguer les poissons du prédateur.


```python
import pygame
import numpy as np
import random

# Paramètres de la simulation
WIDTH, HEIGHT = 800, 600
NUM_FISH = 50
NUM_PREDATOR = 1
SPEED = 2
AVOID_RADIUS = 30
ALIGN_RADIUS = 50
CENTER_RADIUS = 200

# Initialiser Pygame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Simulation de Banc de Poissons")

# Classes
class Fish:
    def __init__(self, x, y):
        self.pos = np.array([x, y], dtype=np.float64)  # Assurez-vous que la position est en float64
        self.vel = np.random.rand(2) * 2 - 1  # Vitesse aléatoire
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED

    def move(self, fishes, predator):
        center_of_mass = np.zeros(2)
        avg_velocity = np.zeros(2)
        avoid_force = np.zeros(2)

        # Règles d'interaction avec les autres poissons
        for fish in fishes:
            if np.linalg.norm(fish.pos - self.pos) < AVOID_RADIUS:
                avoid_force += self.pos - fish.pos
            if np.linalg.norm(fish.pos - self.pos) < ALIGN_RADIUS:
                avg_velocity += fish.vel
            if np.linalg.norm(fish.pos - self.pos) < CENTER_RADIUS:
                center_of_mass += fish.pos

        # Règle du centre du banc
        center_of_mass /= len(fishes)
        move_towards_center = center_of_mass - self.pos

        # Combinaison des forces
        self.vel += (move_towards_center * 0.05 + avg_velocity * 0.1 + avoid_force * 0.1)
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED  # Normaliser la vitesse

        # Déplacement
        self.pos += self.vel

        # Se déplacer dans la fenêtre
        self.pos = self.pos % np.array([WIDTH, HEIGHT])

    def draw(self):
        # Convertir la position en int pour le dessin
        pygame.draw.circle(screen, (0, 255, 0), self.pos.astype(int), 5)

class Predator:
    def __init__(self, x, y):
        self.pos = np.array([x, y], dtype=np.float64)
        self.vel = np.random.rand(2) * 2 - 1
        self.vel = self.vel / np.linalg.norm(self.vel) * SPEED * 1.5  # Prédateur plus rapide

    def move(self, fishes):
        closest_fish = None
        min_dist = float('inf')

        # Chercher le poisson le plus proche
        for fish in fishes:
            dist = np.linalg.norm(fish.pos - self.pos)
            if dist < min_dist:
                min_dist = dist
                closest_fish = fish

        # Se déplacer vers le poisson le plus proche
        if closest_fish:
            move_towards_fish = closest_fish.pos - self.pos
            self.vel = move_towards_fish / np.linalg.norm(move_towards_fish) * SPEED * 1.5

        # Déplacement
        self.pos += self.vel
        self.pos = self.pos % np.array([WIDTH, HEIGHT])

    def draw(self):
        # Convertir la position en int pour le dessin
        pygame.draw.circle(screen, (255, 0, 0), self.pos.astype(int), 10)

# Initialiser les poissons et le prédateur
fishes = [Fish(random.randint(0, WIDTH), random.randint(0, HEIGHT)) for _ in range(NUM_FISH)]
predator = Predator(random.randint(0, WIDTH), random.randint(0, HEIGHT))

# Boucle de simulation
running = True
while running:
    screen.fill((0, 0, 0))  # Fond noir
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Déplacer les poissons
    for fish in fishes:
        fish.move(fishes, predator)

    # Déplacer le prédateur
    predator.move(fishes)

    # Dessiner les poissons
    for fish in fishes:
        fish.draw()

    # Dessiner le prédateur
    predator.draw()

    # Mettre à jour l'affichage
    pygame.display.flip()
    pygame.time.Clock().tick(30)

pygame.quit()
```

Ce code implémente une simulation de banc de poissons avec des comportements de base en utilisant Pygame et Numpy. Voici ce que fait chaque partie du code :

## 1. Initialisation et Paramètres
Le programme commence par initialiser les paramètres nécessaires à la simulation :
- **Largeur et hauteur de la fenêtre** (WIDTH, HEIGHT), ici 800x600 pixels.
- **Nombre de poissons** (NUM_FISH) et **nombre de prédateurs** (NUM_PREDATOR), avec 50 poissons et 1 prédateur.
- **Vitesse des poissons** (SPEED), ainsi que des paramètres pour la distance d'évitement, d'alignement et de cohésion des poissons.

## 2. Initialisation de Pygame
Le module Pygame est initialisé pour créer une fenêtre graphique de taille définie (WIDTH, HEIGHT), avec un titre pour la fenêtre de la simulation.

## 3. Classes : Fish et Predator

### Fish (Poisson) :
- Chaque poisson a une **position** (x, y), une **vitesse** aléatoire et une **direction** qui dépendent de la position actuelle et des voisins.
- **Comportement du poisson** : Lorsqu'un poisson se déplace, il prend en compte :
  - **Cohésion** : Il se déplace vers le centre de masse des autres poissons proches.
  - **Séparation** : Il évite les autres poissons qui sont trop proches.
  - **Alignement** : Il tente de s'aligner avec les autres poissons proches.
- Le poisson se déplace à chaque itération en mettant à jour sa position.
- Lors du rendu, il est dessiné sous forme d'un cercle vert sur l'écran.

### Predator (Prédateur) :
- Le prédateur a aussi une **position**, une **vitesse**, et il se déplace plus vite que les poissons.
- **Comportement du prédateur** : Il cherche le poisson le plus proche et se déplace vers lui.
- Lors du rendu, il est dessiné sous forme d'un cercle rouge sur l'écran.

## 4. Simulation et Boucle Principale
Le programme commence une boucle de simulation où il :
- Écoute les événements (comme la fermeture de la fenêtre avec `pygame.QUIT`).
- Déplace chaque poisson et le prédateur en appliquant leurs règles de comportement respectives.
- Dessine les poissons et le prédateur à chaque itération avec `pygame.draw.circle()`.
- Met à jour l'affichage avec `pygame.display.flip()` et régule la vitesse du jeu avec `pygame.time.Clock().tick(30)`, c'est-à-dire 30 images par seconde.

## 5. Fin de la Simulation
Lorsque l'utilisateur ferme la fenêtre (événement `pygame.QUIT`), la boucle de simulation se termine, et le programme appelle `pygame.quit()` pour fermer proprement Pygame.


- **Poissons** : Ils suivent des comportements de cohésion, séparation et alignement dans un banc pour simuler leur interaction.
- **Prédateur** : Il chasse les poissons en se dirigeant vers le plus proche.

- **Régénération :** les algues se régénèrent à un rythme donné en fonction des conditions de l'écosystème.

### c. Les Herbivores :
Les herbivores marins mangent les algues, et leur population peut augmenter ou diminuer selon la quantité d’algues disponibles et les ressources de leur habitat.

**Propriétés :**
- **Population :** nombre d'herbivores présents.
- **Consommation d'algues :** quantité d'algues mangées par unité de temps.
- **Croissance :** dépend de la disponibilité des algues et des conditions environnementales.

## 2. Définition des interactions :

### a. Interaction entre herbivores et algues :
Les herbivores consomment les algues. Si les herbivores mangent trop d'algues, celles-ci sont appauvries, ce qui peut nuire à la régénération des coraux. La vitesse à laquelle les herbivores consomment les algues peut dépendre de facteurs comme la densité de la population d'herbivores ou la quantité d'algues disponibles.

### b. Interaction entre algues et coraux :
Si la quantité d'algues est suffisante, les coraux se régénèrent plus rapidement. Si les algues sont rares (à cause de la surconsommation par les herbivores ou d'autres facteurs), les coraux souffrent et peuvent mourir. Le taux de régénération des coraux dépend de l'abondance des algues et de l'intensité de la pression exercée par les herbivores.

### c. Pression des herbivores sur les coraux :
Si les herbivores sont trop nombreux ou mangent trop d'algues, cela peut entraîner une surconsommation des ressources, et les coraux ne pourront pas se régénérer suffisamment, ce qui peut mener à leur déclin.

## 3. Modélisation des dynamiques :

### a. Modélisation de la croissance des algues :
Les algues croissent à une vitesse \( G_{\text{algues}} \) en fonction des ressources disponibles et de la température de l'eau. Elles sont consommées à un taux \( C_{\text{herbivores}} \) par les herbivores. Une équation simple pour la quantité d'algues \( A(t) \) à un instant \( t \) pourrait être :
$$
\frac{dA}{dt} = G_{\text{algues}} - C_{\text{herbivores}} \times A(t)\times P_{\text{herbivores}}(t)
$$

### b. Modélisation de la régénération des coraux :
Les coraux se régénèrent à un taux \( R_{\text{coraux}} \) en fonction des algues disponibles, mais cela diminue si les herbivores consomment trop d'algues. L’équation pour la santé des coraux \( S_{\text{coraux}}(t) \) pourrait être :
$$
\frac{dS_{\text{coraux}}}{dt} = R_{\text{coraux}} \times A(t) - \text{pression des herbivores}
$$

### c. Dynamique de la population des herbivores :
La population d'herbivores \( P_{\text{herbivores}}(t) \) varie en fonction de la quantité d'algues disponibles et de la température de l'eau. Une possible dynamique est :
$$
\frac{dP_{\text{herbivores}}}{dt} = r \times P_{\text{herbivores}}(t) \times A(t) - d \times P_{\text{herbivores}}(t)
$$
où :
- \( r \) est le taux de reproduction des herbivores.
- \( d \) est le taux de mortalité des herbivores.
- \( A(t) \) est la quantité d'algues disponibles, influençant la croissance de la population des herbivores.

## Exercice 5 : Simulation du Changement Climatique sur un Écosystème Marin

### Énoncé :
Modéliser un écosystème marin affecté par un changement climatique, avec un accent sur l’acidification des océans et l'impact sur les espèces marines.

### Tâches :
1. **Créer un modèle** qui intègre des agents pour différentes espèces (poissons, coraux, phytoplancton) et l'environnement.

3. **Observer et visualiser** l'impact du changement climatique sur les populations et les interactions entre les espèces.

---

### Description du Modèle :
1. **Les agents** :
   - **Poissons :** Les poissons dépendent des coraux et du phytoplancton pour leur alimentation et leur habitat.
   - **Coraux :** Les coraux, sensibles à l'acidification de l'eau, voient leur capacité à se régénérer affectée par la baisse du pH.
   - **Phytoplancton :** Le phytoplancton, qui constitue la base de la chaîne alimentaire marine, est influencé par la température de l'eau et l'acidité.

2. **Le changement climatique :**
   - Un facteur de **baisse du pH** de l'eau simulera l'acidification des océans. Cette baisse du pH affectera la santé des coraux, perturbant ainsi les interactions avec les autres espèces.
   
3. **Interactivité :**
   - Les poissons se nourrissent principalement de phytoplancton et vivent dans les récifs coralliens.
   - L'acidification de l'eau (réduisant le pH) affecte la régénération des coraux, perturbant ainsi l'équilibre de l'écosystème.
   - Le modèle prendra en compte l'impact du changement climatique sur la croissance du phytoplancton et l'alimentation des poissons.

4. **Simulation et visualisation :**
   - Nous allons simuler l'évolution de l'écosystème marin sur une période donnée, en observant les variations dans les populations d'espèces et la qualité de l'environnement.
   - Les résultats seront visualisés à l’aide de graphiques qui montreront l’évolution des populations de poissons, coraux et phytoplancton sous l'impact du changement climatique.

---

### Outils à utiliser :
- **Python avec des bibliothèques telles que matplotlib et numpy** pour effectuer les simulations et visualiser les résultats.
- **Bibliothèques de simulation basées sur des agents** comme NetLogo (si besoin de visualiser les interactions d'agents plus complexes).

Voici l'intégralité du code de la simulation du changement climatique sur un écosystème marin dans un seul bloc, avec des commentaires pour chaque étape du processus. Ce code simule l'évolution des populations de phytoplancton, coraux et poissons sous l'impact de l'acidification des océans.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Paramètres de simulation ---
timestep = 1  # Une unité de temps (par exemple, années)
num_years = 100  # Durée de la simulation (en années)

# Paramètres du changement climatique
initial_pH = 8.1  # pH initial de l'eau
acidification_rate = 0.01  # Changement du pH par an (acidification)

# Populations initiales
initial_phytoplankton = 1000  # Nombre initial de phytoplanctons
initial_corals = 500  # Nombre initial de coraux
initial_fish = 300  # Nombre initial de poissons

# --- Classes pour les agents ---
# 1. Phytoplancton : Croît en fonction du pH et de la disponibilité des nutriments.
class Phytoplankton:
    def __init__(self, population):
        self.population = population

    def grow(self, pH):
        # Le phytoplancton croît plus lentement à un pH plus bas
        growth_rate = 0.1 if pH > 7.8 else 0.05  # Croissance plus lente si l'acidification progresse
        self.population += growth_rate * self.population

# 2. Coraux : Dépendent du pH pour croître et se reproduire.
class Coral:
    def __init__(self, population):
        self.population = population

    def grow(self, pH):
        # Croissance des coraux dépendant du pH
        if pH < 7.8:
            death_rate = 0.1  # Augmentation du taux de mortalité avec l'acidification
        else:
            death_rate = 0.02  # Croissance normale
        self.population *= (1 - death_rate)

# 3. Poissons : Se nourrissent de phytoplancton et sont sensibles à la baisse du pH.
class Fish:
    def __init__(self, population):
        self.population = population

    def feed_and_reproduce(self, phytoplankton_population, pH):
        # Les poissons se nourrissent du phytoplancton et se reproduisent en fonction de sa quantité
        if pH < 7.5:  # Stress dû à l'acidification
            survival_rate = 0.5  # Moins de reproduction et plus de mortalité
        else:
            survival_rate = 0.9  # Taux de survie plus élevé dans de meilleures conditions
        # Les poissons mangent du phytoplancton, donc leur population dépend de cette source de nourriture
        self.population += (0.05 * phytoplankton_population * survival_rate)

# --- Initialisation des agents ---
# Initialisation des populations de chaque espèce
phytoplankton = Phytoplankton(initial_phytoplankton)
corals = Coral(initial_corals)
fish = Fish(initial_fish)

# Liste pour suivre les résultats au cours des années
results = []

# --- Simulation sur plusieurs années ---
for year in range(num_years):
    # Calcul du pH à chaque année (avec acidification)
    pH = initial_pH - acidification_rate * year
    
    # Les agents interagissent selon le pH actuel
    phytoplankton.grow(pH)
    corals.grow(pH)
    fish.feed_and_reproduce(phytoplankton.population, pH)
    
    # Enregistrer l'état actuel de l'écosystème
    results.append([year, pH, phytoplankton.population, corals.population, fish.population])

# --- Conversion des résultats en DataFrame ---
df = pd.DataFrame(results, columns=["Year", "pH", "Phytoplankton", "Corals", "Fish"])
df.set_index("Year", inplace=True)

# Affichage des premières lignes du tableau pour vérifier les résultats
df.head()

# --- Visualisation des résultats ---

# Créer une figure pour afficher les graphiques
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Graphique 1: Évolution du pH de l'eau
ax[0].plot(df.index, df["pH"], color="blue", label="pH")
ax[0].set_title("Évolution du pH de l'océan")
ax[0].set_xlabel("Année")
ax[0].set_ylabel("pH")
ax[0].grid(True)

# Graphique 2: Évolution des populations de Phytoplancton, Coraux et Poissons
ax[1].plot(df.index, df["Phytoplankton"], label="Phytoplancton", color="green")
ax[1].plot(df.index, df["Corals"], label="Coraux", color="orange")
ax[1].plot(df.index, df["Fish"], label="Poissons", color="red")
ax[1].set_title("Évolution des populations marines")
ax[1].set_xlabel("Année")
ax[1].set_ylabel("Population")
ax[1].legend()
ax[1].grid(True)

# Ajuster la mise en page pour une meilleure présentation
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Paramètres de simulation ---
timestep = 1  # Une unité de temps (par exemple, années)
num_years = 100  # Durée de la simulation (en années)

# Paramètres de la température SST
initial_SST = 25  # Température initiale de surface de la mer (en °C)
warming_rate = 0.1  # Augmentation annuelle de la température SST (en °C par an)

# Populations initiales
initial_phytoplankton = 1000  # Nombre initial de phytoplanctons
initial_corals = 500  # Nombre initial de coraux
initial_fish = 300  # Nombre initial de poissons

# --- Classes pour les agents ---
# 1. Phytoplancton : Croît en fonction de la température de la mer.
class Phytoplankton:
    def __init__(self, population):
        self.population = population

    def grow(self, SST):
        # Le phytoplancton croît mieux à des températures modérées
        if SST > 28:
            growth_rate = -0.05  # Croissance ralentie si SST trop élevée
        elif SST < 18:
            growth_rate = 0.05  # Croissance optimale à des températures modérées
        else:
            growth_rate = 0.1  # Croissance optimale à des températures modérées (21-28°C)
        self.population += growth_rate * self.population

# 2. Coraux : Les coraux subissent un stress accru avec la température élevée (blanchissement).
class Coral:
    def __init__(self, population):
        self.population = population

    def grow(self, SST):
        # Les coraux subissent un stress si la SST dépasse un certain seuil
        if SST > 30:
            death_rate = 0.1  # Blanchissement des coraux (perte de population)
        elif SST > 28:
            death_rate = 0.05  # Stress léger
        else:
            death_rate = 0.02  # Croissance normale
        self.population *= (1 - death_rate)

# 3. Poissons : Dépendent de la disponibilité du phytoplancton et sont sensibles à la température.
class Fish:
    def __init__(self, population):
        self.population = population

    def feed_and_reproduce(self, phytoplankton_population, SST):
        # Les poissons souffrent avec une température trop élevée
        if SST > 30:  # Stress intense avec SST élevée
            survival_rate = 0.5  # Moins de reproduction et plus de mortalité
        else:
            survival_rate = 0.9  # Croissance normale
        # Les poissons se nourrissent du phytoplancton, donc leur population dépend de cette ressource
        self.population += (0.05 * phytoplankton_population * survival_rate)

# --- Initialisation des agents ---
# Initialisation des populations de chaque espèce
phytoplankton = Phytoplankton(initial_phytoplankton)
corals = Coral(initial_corals)
fish = Fish(initial_fish)

# Liste pour suivre les résultats au cours des années
results = []

# --- Simulation sur plusieurs années ---
for year in range(num_years):
    # Calcul de la température SST à chaque année (avec réchauffement annuel)
    SST = initial_SST + warming_rate * year
    
    # Les agents interagissent selon la température SST actuelle
    phytoplankton.grow(SST)
    corals.grow(SST)
    fish.feed_and_reproduce(phytoplankton.population, SST)
    
    # Enregistrer l'état actuel de l'écosystème
    results.append([year, SST, phytoplankton.population, corals.population, fish.population])

# --- Conversion des résultats en DataFrame ---
df = pd.DataFrame(results, columns=["Year", "SST", "Phytoplankton", "Corals", "Fish"])
df.set_index("Year", inplace=True)

# Affichage des premières lignes du tableau pour vérifier les résultats
df.head()

# --- Visualisation des résultats ---

# Créer une figure pour afficher les graphiques
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Graphique 1: Évolution de la température SST
ax[0].plot(df.index, df["SST"], color="red", label="SST (°C)")
ax[0].set_title("Évolution de la Température de Surface de la Mer (SST)")
ax[0].set_xlabel("Année")
ax[0].set_ylabel("Température SST (°C)")
ax[0].grid(True)

# Graphique 2: Évolution des populations de Phytoplancton, Coraux et Poissons
ax[1].plot(df.index, df["Phytoplankton"], label="Phytoplancton", color="green")
ax[1].plot(df.index, df["Corals"], label="Coraux", color="orange")
ax[1].plot(df.index, df["Fish"], label="Poissons", color="blue")
ax[1].set_title("Évolution des populations marines")
ax[1].set_xlabel("Année")
ax[1].set_ylabel("Population")
ax[1].legend()
ax[1].grid(True)

# Ajuster la mise en page pour une meilleure présentation
plt.tight_layout()
plt.show()
```

## Explication des Modifications

### 1. **Température de Surface de la Mer (SST) :**
Nous remplaçons l'effet de la concentration de **CO₂** par une **augmentation progressive de la température SST**. Chaque année, la température SST augmente selon un taux de réchauffement (`warming_rate`) fixe.

### 2. **Effets sur le Phytoplancton :**
Le phytoplancton croît plus lentement lorsque la température SST dépasse des seuils critiques. Les températures entre **18°C** et **28°C** sont idéales pour leur croissance, mais au-dessus de **28°C**, leur croissance est freinée.

### 3. **Impact sur les Coraux :**
Les **coraux** sont très sensibles aux températures élevées. À mesure que la température de la mer augmente au-delà de **28°C**, ils subissent un stress, et à des températures supérieures à **30°C**, un **blanchissement des coraux** est simulé, réduisant leur population.

### 4. **Impact sur les Poissons :**
Les **poissons** dépendent de la disponibilité du **phytoplancton**. Si la température augmente trop, la population de poissons est affectée par le stress thermique et la diminution de leur source de nourriture. Leur survie et reproduction sont réduites.
