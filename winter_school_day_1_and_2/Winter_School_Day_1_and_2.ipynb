### Branching-model

- This code models the exponential growth of a branch population, where the growth rate is proportional to the current population. The solution to the differential equation governing this growth is computed numerically and plotted over a 10-unit time period.

```python
import numpy as np            
import matplotlib.pyplot as plt  
from scipy.integrate import odeint  

# Model parameters
b = 0.25  # Decay constant, influences the rate of growth of the number of branches

# Initial condition
n0 = 10  # Initial number of branches

# Definition of the differential equation describing the evolution of n(t)
def dndt(n, t):
    return b * n

# Creating a time array from 0 to 10, with 101 points
t_array = np.linspace(0, 10, 101)

# Solving the differential equation using odeint
n = odeint(dndt, n0, t_array)

# Visualization
plt.plot(t_array, n, linewidth=3)
plt.title("Branching Model")
plt.xlabel("Time [t]", fontweight='bold')
plt.ylabel("n(t)", fontweight='bold')
plt.xlim([0,10])
plt.grid(True)
plt.savefig("fig_branching_model.png")
plt.show()
```

### Cat and mouse model
- This code models the population dynamics of a species, considering birth, death, and decay rates. The population changes over time based on these factors, and the solution to the differential equation is computed numerically and visualized in a plot over a 10-unit time period.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
b = 0.25  # decay constant
m = 0.2   # birth rate for population
d = 0.5   # death rate for population

# Initial condition: Starting with 10 individuals in the population
n0 = 10  

# dndt
def dndt(n,t):
    return b*n-d*n+m
t_array = np.linspace(0,10,101)
n=odeint(dndt,n0,t_array)

# Visualization
plt.plot(t_array, n, linewidth='3')
plt.title("Cat and mouse model")
plt.xlabel("Time [t]", fontweight='bold')
plt.ylabel("n(t)", fontweight='bold')
plt.xlim([0,10])
plt.grid(True)
plt.savefig("fig_Cat_and_mouse_model.png")        
plt.show()
```

### Flu model
- This code models the evolution of healthy individuals and carriers in an epidemic scenario. The transmission rate and the number of contacts per day determine the rate of change of both groups over time. The solution is computed numerically and visualized with two y-axes showing the dynamics of healthy individuals, carriers, and the total population over a 10-day period.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
a, c = 0.025, 10  # a = Transmission rate (probability of 2.5%), c = Number of contacts per day

def model(y, t, a, c):
    s, n = y
    ds_dt = -a * c * n * s  # Rate of change of healthy individuals
    dn_dt = a * c * n * s   # Rate of change of carriers
    return [ds_dt, dn_dt]

# Initial conditions
s0, n0 = 95.0, 5.0  # s0 = Initial number of healthy individuals, n0 = Initial number of carriers
y0 = [s0, n0]
t = np.linspace(0, 10, 101)  # Time period (in days)
params = (a, c)
solution = odeint(model, y0, t, args=params)

s, n = solution[:, 0], solution[:, 1]

# Plot results with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot s(t) on the first axis (blue)
ax1.plot(t, s, '-b', label='Healthy individuals s(t)', linewidth=3)
ax1.set_xlim(0, 10)
ax1.set_ylim(0, max(s) * 1.1)  # Dynamically adjust the limits for s(t)
ax1.set_xlabel('Time steps', fontsize=14, fontweight='bold')
ax1.set_ylabel('Healthy individuals', color='blue', fontsize=14)
ax1.tick_params(axis='y', colors='blue')

# Create a second y-axis for the carriers
ax2 = ax1.twinx()

# Plot n(t) on the second axis (red)
ax2.plot(t, n, '-r', label='Carriers n(t)', linewidth=3)
ax2.set_ylabel('Carriers', color='red', fontsize=14)
ax2.tick_params(axis='y', colors='red')
ax2.set_ylim(0, max(n) * 1.1)  # Dynamically adjust the limits for n(t)
ax2.set_xlim(0, 10)

# Add the total population
ax1.plot(t, s + n, '--g', label='Total population s(t) + n(t)', linewidth=3)
plt.title('Evolution of healthy individuals, carriers, and total population', fontsize=16)
plt.grid(True)
ax1.legend(loc='best', fontsize=12)
plt.savefig("fig_Flu_model.png")        
plt.show()
```

# Predator-Prey Interactions

To illustrate the Lotka-Volterra predation model with time series in Python, we can use differential equations to model the populations of prey and predators. Here is an example of code that simulates and visualizes the results of prey and predator populations over time.

The Lotka-Volterra differential equations are as follows:

$$
\frac{dx}{dt} = \alpha x - \beta x y \quad \text{(prey growth rate)}
$$

$$
\frac{dy}{dt} = \delta x y - \gamma y \quad \text{(predator growth rate)}
$$

Where:

- \( x \) is the prey population,
- \( y \) is the predator population,
- \( \alpha \), \( \beta \), \( \gamma \), and \( \delta \) are parameters.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
alpha, beta, delta, gamma = 0.001, 0.05, 0.03, 0.0002  # Prey growth rate, predation rate, predator growth rate, predator mortality rate

def model(z, t, alpha, beta, delta, gamma):
    x, y = z
    dx_dt = alpha * x - beta * x * y
    dy_dt = delta * x * y - gamma * y
    return [dx_dt, dy_dt]

x0, y0 = 160, 3
z0 = [x0, y0]
t = np.linspace(0, 10, 101)  # 10-year period
params = (alpha, beta, delta, gamma)
solution = odeint(model, z0, t, args=params)

x, y = solution[:, 0], solution[:, 1]

# Plotting the results
ax1 = plt.subplot(1, 1, 1)
ax1.plot(t, x, '-b', label='Prey', linewidth=3)
ax1.set_xlim(0, 10)
ax1.set_ylim(0, max(x) * 1.1)  # Dynamically adjust limits
ax1.set_xlabel('Time steps', fontsize=20, fontweight='bold')
ax1.set_ylabel('Prey population', color='blue', fontsize=20)
ax1.tick_params(axis='y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(t, y, '-r', label='Predators', linewidth=3)
ax2.set_ylabel('Predator population', color='red', fontsize=20)
ax2.tick_params(axis='y', colors='red')
ax2.set_ylim(0, max(y) * 1.1)  # Dynamically adjust limits
ax2.set_xlim(0, 10)

# Add legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Predator-prey dynamics')
plt.grid(True)
plt.savefig("fig_predator_prey_model_model.png")        
plt.show()
```

## predator-prey model
- This code simulates the dynamics between a predator and prey population over time using a set of differential equations. The growth of the prey population and the predation rate influence the predator population. The solution is numerically computed and visualized with two y-axes showing the predator and prey populations over a 150-unit time period.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
r, a, c, epsilon, delta = 0.5, 1.0, 0.01, 0.02, 0.1  # Prey growth rate, Predation rate, Predator growth rate, Predator death rate

def model(n, t, r, a, c, epsilon, delta):
    n1, n2 = n
    dn1_dt = r * n1 - a * c * n1 * n2  # Rate of change for prey population
    dn2_dt = epsilon * a * c * n1 * n2 - delta * n2  # Rate of change for predator population
    return [dn1_dt, dn2_dt]

# Initial conditions
n10, n20 = 500, 30  # n10 = Initial prey population, n20 = Initial predator population
n0 = [n10, n20]
t = np.linspace(0, 150, 1001)  # Time period (in time steps)
params = (r, a, c, epsilon, delta)
solution = odeint(model, n0, t, args=params)

n1, n2 = solution[:, 0], solution[:, 1]

# Plotting the results with two y-axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot prey population on the first axis (red)
ax1.plot(t, n1, '-r', label='Prey', linewidth=3)
ax1.set_xlim(0, 150)
ax1.set_ylim(0,)  # Adjust dynamically for the prey population
ax1.set_xlabel('Time steps', fontsize=20, fontweight='bold')
ax1.set_ylabel('Prey Population', color='red', fontsize=20)
ax1.tick_params(axis='y', colors='red')

# Create a second y-axis for the predator population
ax2 = ax1.twinx()

# Plot predator population on the second axis (blue)
ax2.plot(t, n2, '-b', label='Predator', linewidth=3)
ax2.set_ylabel('Predator Population', color='blue', fontsize=20)
ax2.tick_params(axis='y', colors='blue')
ax2.set_ylim(0,)   # Adjust dynamically for the predator population
ax2.set_xlim(0, 150)

# Adding a single combined legend
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

# Title and grid
plt.title('Predator-Prey Dynamics', fontsize=16)
plt.grid(True)
plt.savefig("fig_predator-prey_model_model.png")        
# Show the plot
plt.show()
```

### PZN Model
- This code simulates the interactions between phytoplankton, zooplankton, and nutrients in an ecosystem using a set of differential equations. The model incorporates growth, grazing, mortality, and nutrient recycling. The populations of phytoplankton, zooplankton, and nutrients are solved numerically and visualized over a 150-unit time period.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
umax, Kn, gmax, v, p0, mp, mz, gama = 2.0, 1.0, 1.5, 1.0, 0.0, 0.1, 0.2, 0.3  # These are the parameters of the model
# - umax: Maximum growth rate of phytoplankton (P)
# - Kn: Half-saturation constant for nutrients (N)
# - gmax: Maximum grazing rate of zooplankton (Z)
# - v: Saturation constant for the grazing function
# - p0: A threshold value for phytoplankton, below which grazing decreases
# - mp: Mortality rate of phytoplankton
# - mz: Mortality rate of zooplankton
# - gama: Proportion of nutrients that are recycled after zooplankton death

def model(Y, t, umax, Kn, gmax, v, p0, mp, mz, gama):
    """
    Defines the system of ODEs for the PZN model (Phytoplankton-Zooplankton-Nutrient).
    Y = [P, Z, N] where P, Z, N are the populations of phytoplankton, zooplankton, and nutrients, respectively.
    """
    P, Z, N = Y
    dP_dt = umax * (N / (N + Kn)) * P - mp * P - gmax * (1 - np.exp(-v * (P - p0))) * Z
    dZ_dt = gama * gmax * (1 - np.exp(-v * (P - p0))) * Z - mz * Z 
    dN_dt = -umax * (N / (N + Kn)) * P + mp * P + mz * Z + (1 - gama) * gmax * (1 - np.exp(-v * (P - p0))) * Z
    return [dP_dt, dZ_dt, dN_dt]

# Initial conditions for phytoplankton (P), zooplankton (Z), and nutrients (N)
P0, Z0, N0 = 0.3, 0.1, 1.6
Y0 = [P0, Z0, N0]  
t = np.linspace(0, 150, 1001) 

params = (umax, Kn, gmax, v, p0, mp, mz, gama)  # Grouping all model parameters into a tuple

# Solve the system of ODEs using odeint (numerical integration)
solution = odeint(model, Y0, t, args=params)

# Extract the solution for each population (phytoplankton, zooplankton, and nutrients)
P, Z, N = solution[:, 0], solution[:, 1], solution[:, 2]

# Plotting the results
plt.figure(figsize=(12, 6))  # Set figure size for clarity

# Plot the phytoplankton population (P) vs time
plt.plot(t, P, label='Phytoplankton Population', color='green')
plt.plot(t, Z, label='Zooplankton Population', color='blue')
plt.plot(t, N, label='Nutrient Population', color='red')
plt.xlim(0, 150)

# Set dynamic y-axis limits based on the maximum value across all populations
max_y = max([max(P), max(Z), max(N)])  
plt.ylim(0.1, max_y)  

# Add labels, title, and grid
plt.title('PZN Model')  
plt.xlabel('Time steps', fontweight='bold') 
plt.ylabel('Population Size', fontweight='bold')
plt.legend()  
plt.grid(True)  
plt.savefig("fig_PZN_model_model.png")        
plt.show()
```

### PZN Model with different parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def model(Y, t, umax, Kn, gmax, v, p0, mp, mz, gama):
    P, Z, N = Y
    dP_dt = umax * (N / (N + Kn)) * P - mp * P - gmax * (1 - np.exp(-v * (P - p0))) * Z
    dZ_dt = gama * gmax * (1 - np.exp(-v * (P - p0))) * Z - mz * Z
    dN_dt = -umax * (N / (N + Kn)) * P + mp * P + mz * Z + (1 - gama) * gmax * (1 - np.exp(-v * (P - p0))) * Z
    return [dP_dt, dZ_dt, dN_dt]

# Function to solve the model and plot the results
def solve_and_plot(umax, Kn, gmax, v, p0, mp, mz, gama, P0, Z0, N0, ax):
    Y0 = [P0, Z0, N0]
    t = np.linspace(0, 150, 1001)  # Time vector
    params = (umax, Kn, gmax, v, p0, mp, mz, gama)

    # Solve the ODE system
    solution = odeint(model, Y0, t, args=params)

    # Extract results
    P, Z, N = solution[:, 0], solution[:, 1], solution[:, 2]

    # Plotting the results
    ax.plot(t, P, label='Phytoplankton Population', color='green')
    ax.plot(t, Z, label='Zooplankton Population', color='blue')
    ax.plot(t, N, label='Nutrient Population', color='red')
    ax.set_title('PZN Model')
    ax.set_xlabel('Time steps', fontweight='bold')
    ax.set_ylabel('Population Size', fontweight='bold')
    ax.legend()
    ax.grid()

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5))

# First simulation
solve_and_plot(2.0, 1.0, 1.5, 1.0, 0.0, 0.1, 0.2, 0.3, 0.3, 0.1, 1.6, ax1)

# Second simulation
solve_and_plot(0.69, 1.0, 1.5, 1.0, 0.0, 0.1, 0.2, 0.3, 0.3, 0.1, 10.6, ax2)

# Third simulation
solve_and_plot(0.69, 1.0, 1.5, 1.0, 1.0, 0.1, 0.2, 0.3, 0.3, 0.1, 10.6, ax3)

# Adjust layout and show the plots
plt.tight_layout()
plt.savefig("fig_PZN_models.png")        
plt.show()
```
