# Energy Balance Model

This code models the evolution of the temperature of an aquatic system over time using an energy balance model. It calculates the evolution of the water's thermal content based on various physical parameters such as solar luminosity, albedo, emissivity, and the heat capacity of the water. 

The temperature evolution is described by a differential equation that takes into account the energy entering the system (e.g., sunlight) and the energy leaving the system (e.g., heat emission to space). 

The code solves this equation for several initial temperature conditions and displays the results in a graph, showing how the water temperature changes over time.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Model parameters
timeStep = 1  # in years
waterDepth = 4000  # in meters
L = 1350  # solar luminosity in W/m^2
alpha = 0.3  # albedo
epsilon = 1  # emissivity
sigma = 5.67e-8  # Stefan-Boltzmann constant in W/m^2 K^4
s_y = 3.14e7  # seconds per year
heatcapacity = waterDepth * 4.2e6  # J/K m^2

# Initial conditions: temperatures in K
T_init = [0, 100, 150, 288, 303, 500] 
# Initial heat content in J/m^2
heatContent_init = np.array(heatcapacity * np.array(T_init))  

# Definition of the differential equation describing the evolution of heat content
def dheatcontent_dt(heatcontent, t):
    # Temperature (K) derived from heat content
    T = heatcontent / heatcapacity  
    # Energy balance equation
    dQdt = ((L * (1 - alpha)) / 4 - epsilon * sigma * (T**4)) * s_y  # Energy input - Energy output
    return dQdt

# Creating a time array from 0 to 500 years, with 1000 points
t = np.linspace(0, 2000, 1000)

# Solving the differential equation using odeint for each initial condition
heatContent_all = []
for initial_heat_content in heatContent_init:
    heatContent = odeint(dheatcontent_dt, initial_heat_content, t)
    heatContent_all.append(heatContent)

# Visualization: Plotting temperature (heat content / heatcapacity) over time for each initial condition
plt.figure(figsize=(10, 6))
for heatContent in heatContent_all:
    plt.plot(t, heatContent / heatcapacity, linewidth=2)

plt.title("Energy Balance Model")
plt.xlabel("Time (years)", fontsize=14, fontweight='bold')
plt.ylabel("Temperature (K)",fontsize=14, fontweight='bold')
plt.grid(True)
plt.legend([f"Initial Temp = {T} K" for T in T_init], loc='best')
plt.savefig("fig1.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
timestep = 1  # in years
waterdepth = 4000  # in meters
L = 1350  # solar luminosity in W/m^2
epsilon = 1  # emissivity
sigma = 5.67e-8  # Stefan-Boltzmann constant in W/m^2 K^4
s_y = 3.14e7  # seconds per year
heatcapacity = waterdepth * 4.2e6  # J/K m^2

# Initial heat content
h0 = 0.0  # Initial heat content

# Function to calculate heat content based on albedo
def func(h, t, a):
    T = h / heatcapacity  # Temperature derived from heat content
    dHdt = ((L * (1 - a) / 4) - epsilon * sigma * (T**4)) * s_y  # Energy balance equation
    return dHdt

# Albedo values from 0 to 1
albedo_values = np.arange(0, 1.01, 0.01)

# Initialize a list to store heat content values for each albedo
heat_content = []

# Solve the differential equation for each albedo
for a in albedo_values:
    h = odeint(func, h0, np.linspace(0, 2000, 1000), args=(a,))
    heat_content.append(h[-1, 0])  # Take the final heat content after 2000 years

# Plot heat content over different albedo values
plt.figure(figsize=(10, 6))
plt.plot(albedo_values, heat_content, label='Heat content vs Albedo')
plt.xlabel('Albedo',fontsize=14, fontweight='bold')
plt.ylabel('Heat Content (J/m^2)',fontsize=14, fontweight='bold')
plt.title('Heat Content vs Albedo')
plt.axhline(0.428e13, color='k', label='Steady state normal parameters')
plt.legend()
plt.grid(True)
plt.savefig("fig5.png")
plt.show()
```
